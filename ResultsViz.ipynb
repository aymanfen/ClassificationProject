```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Metrics.csv",index_col=0)
df.head()
```

```python
#Create plot
df=df.sort_values('precision')
plt.figure(figsize=(8,6))
plt.scatter(df['precision'],df['model'],color='red')
plt.grid(True)

# Add labels and title
plt.xlabel('Precision')
plt.ylabel('Model')
plt.title('Precision Values by Model')

#show the plot
plt.show()
```

```python
#Create plot
df=df.sort_values('recall')
plt.figure(figsize=(8,6))
plt.scatter(df['recall'],df['model'],color='red')
plt.grid(True)

# Add labels and title
plt.xlabel('Recall')
plt.ylabel('Model')
plt.title('Recall Values by Model')

#show the plot
plt.show()
```

```python
#Create plot
df=df.sort_values('f1')
plt.figure(figsize=(8,6))
plt.scatter(df['f1'],df['model'],color='red')
plt.grid(True)

# Add labels and title
plt.xlabel('F1 Score')
plt.ylabel('Model')
plt.title('F1 Values by Model')

#show the plot
plt.show()
```
